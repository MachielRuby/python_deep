# 双门限法

```python

def FrameTimeC(frameNum, frameLen, inc, fs):
    ll = np.array([i for i in range(frameNum)])
    return ((ll - 1) * inc + frameLen / 2) / fs

def STEn(x, win, inc):
    """
    计算短时能量函数
    :param x:
    :param win:
    :param inc:
    :return:
    """
    X = enframe(x, win, inc)
    s = np.multiply(X, X)
    return np.sum(s, axis=1)


def STZcr(x, wlen, inc, delta=0.01):
    """
    计算信号的短时过零率。

    参数：
    - x: numpy数组，输入信号
    - wlen: int，分析窗口的长度（样本数）
    - inc: int，帧移（样本数）
    - delta: float，避免零除法的小常数

    返回：
    - zcr: numpy数组，短时过零率
    """

    def zero_crossings(frame):
        return np.sum(np.abs(np.diff(np.sign(frame)))) / 2

    minlen = 5  # 可靠计算ZCR所需的最小帧数
    status = 0
    y = enframe(x, wlen, inc)
    fn = y.shape[0]
    zcr = np.zeros(fn)

    for i in range(fn):
        zcr[i] = zero_crossings(y[i, :])

    zcr = (zcr + delta) / (wlen + delta)

    return zcr
```

```python

def findSegment(express):
    """
    分割成語音段
    :param express:
    :return:
    """
    if express[0] == 0:
        voiceIndex = np.where(express)
    else:
        voiceIndex = express
    d_voice = np.where(np.diff(voiceIndex) > 1)[0]
    voiceseg = {}
    if len(d_voice) > 0:
        for i in range(len(d_voice) + 1):
            seg = {}
            if i == 0:
                st = voiceIndex[0]
                en = voiceIndex[d_voice[i]]
            elif i == len(d_voice):
                st = voiceIndex[d_voice[i - 1]+1]
                en = voiceIndex[-1]
            else:
                st = voiceIndex[d_voice[i - 1]+1]
                en = voiceIndex[d_voice[i]]
            seg['start'] = st
            seg['end'] = en
            seg['duration'] = en - st + 1
            voiceseg[i] = seg
    return voiceseg


def vad_TwoThr(x, wlen, inc, NIS):
    """
    使用门限法检测语音段
    :param x: 语音信号
    :param wlen: 分帧长度
    :param inc: 帧移
    :param NIS:
    :return:
    """
    maxsilence = 70
    minlen = 40
    status = 0
    y = enframe(x, wlen, inc)
    fn = y.shape[0]
    amp = STEn(x, wlen, inc)
    zcr = STZcr(x, wlen, inc)
    ampth = np.mean(amp[:NIS])
    zcrth = np.mean(zcr[:NIS])
    amp2 = 2 * ampth
    amp1 = 4 * ampth
    zcr2 = 2 * zcrth
    xn = 0
    count = np.zeros(fn)
    silence = np.zeros(fn)
    x1 = np.zeros(fn)
    x2 = np.zeros(fn)
    for n in range(fn):
        if status == 0 or status == 1:
            if amp[n] > amp1:
                x1[xn] = max(1, n - count[xn] - 1)
                status = 2
                silence[xn] = 0
                count[xn] += 1
            elif amp[n] > amp2 or zcr[n] > zcr2:
                status = 1
                count[xn] += 1
            else:
                status = 0
                count[xn] = 0
                x1[xn] = 0
                x2[xn] = 0

        elif status == 2:
            if amp[n] > amp2 and zcr[n] > zcr2:
                count[xn] += 1
            else:
                silence[xn] += 1
                if silence[xn] < maxsilence:
                    count[xn] += 1
                elif count[xn] < minlen:
                    status = 0
                    silence[xn] = 0
                    count[xn] = 0
                else:
                    status = 3
                    x2[xn] = x1[xn] + count[xn]
        elif status == 3:
            status = 0
            xn += 1
            count[xn] = 0
            silence[xn] = 0
            x1[xn] = 0
            x2[xn] = 0
    el = len(x1[:xn])
    if x1[el - 1] == 0:
        el -= 1
    if x2[el - 1] == 0:
        print('Error: Not find endding point!\n')
        x2[el] = fn
    SF = np.zeros(fn)
    NF = np.ones(fn)
    for i in range(el):
        SF[int(x1[i]):int(x2[i])] = 1
        NF[int(x1[i]):int(x2[i])] = 0
    voiceseg = findSegment(np.where(SF == 1)[0])
    vsl = len(voiceseg.keys())
    return voiceseg, vsl, SF, NF, amp, zcr

```

```python
import librosa
def enframe(x, win, inc=None):
    nx = len(x)
    if isinstance(win, list):
        nwin = len(win)
        nlen = nwin  # 帧长=窗长
    elif isinstance(win, int):
        nwin = 1
        nlen = win  # 设置为帧长
    if inc is None:
        inc = nlen
    nf = (nx - nlen + inc) // inc
    frameout = np.zeros((nf, nlen))
    indf = np.multiply(inc, np.array([i for i in range(nf)]))
    for i in range(nf):
        frameout[i, :] = x[indf[i]:indf[i] + nlen]
    if isinstance(win, list):
        frameout = np.multiply(frameout, np.array(win))
    return frameout
file_path = '../01.wav'
data, fs = librosa.load(file_path, sr=None)
data /= np.max(data)
N = len(data)
wlen = 200
inc = 80
IS = 0.3
overlap = wlen - inc
NIS = int((IS * fs - wlen) // inc + 1)
fn = (N - wlen) // inc + 1

frameTime = FrameTimeC(fn, wlen, inc, fs)
time = [i / fs for i in range(N)]

voiceseg, vsl, SF, NF, amp, zcr = vad_TwoThr(data, wlen, inc, NIS)
plt.figure(figsize=(12,8))
plt.subplot(3, 1, 1)
plt.plot(time, data)
plt.title("双门限法的端点检测")
plt.ylabel("幅值")

plt.subplot(3, 1, 2)
plt.plot(frameTime, amp)
plt.title("短时能量")
plt.ylabel("幅值")

plt.subplot(3, 1, 3)
plt.plot(frameTime, zcr)
plt.title("短时过零率")
plt.ylabel("幅值")
plt.xlabel("时间/s")
for i in range(vsl):
    plt.subplot(3, 1, 1)
    plt.vlines(x=frameTime[voiceseg[i]['start']], ymin=np.min(data), ymax=np.max(data), colors='b', linestyles='dotted')
    plt.vlines(x=frameTime[voiceseg[i]['end']],ymin=np.min(data), ymax=np.max(data), colors='r', linestyles='solid')
    plt.legend(['signal', 'start', 'end'])
    plt.subplot(3, 1, 2)
    plt.vlines(x=frameTime[voiceseg[i]['start']], ymin=np.min(amp), ymax=np.max(amp), colors='b', linestyles='dotted')
    plt.vlines(x=frameTime[voiceseg[i]['end']], ymin=np.min(amp), ymax=np.max(amp), colors='r', linestyles='solid')
    plt.legend(['signal', 'start', 'end'])
    plt.subplot(3, 1, 3)
    plt.vlines(x=frameTime[voiceseg[i]['start']], ymin=np.min(zcr), ymax=np.max(zcr), colors='b', linestyles='dotted')
    plt.vlines(x=frameTime[voiceseg[i]['end']],ymin=np.min(zcr), ymax=np.max(zcr), colors='r', linestyles='solid')
    plt.legend(['signal', 'start', 'end'])


plt.show()

```

```python

def vad_forw(dst1, T1, T2):
    fn = len(dst1)
    maxsilence = 8
    minlen = 5
    status = 0
    count = np.zeros(fn)
    silence = np.zeros(fn)
    xn = 0
    x1 = np.zeros(fn)
    x2 = np.zeros(fn)
    for n in range(1, fn):
        if status == 0 or status == 1:
            if dst1[n] > T2:
                x1[xn] = max(1, n - count[xn] - 1)
                status = 2
                silence[xn] = 0
                count[xn] += 1
            elif dst1[n] > T1:
                status = 1
                count[xn] += 1
            else:
                status = 0
                count[xn] = 0
                x1[xn] = 0
                x2[xn] = 0
        if status == 2:
            if dst1[n] > T1:
                count[xn] += 1
            else:
                silence[xn] += 1
                if silence[xn] < maxsilence:
                    count[xn] += 1
                elif count[xn] < minlen:
                    status = 0
                    silence[xn] = 0
                    count[xn] = 0
                else:
                    status = 3
                    x2[xn] = x1[xn] + count[xn]
        if status == 3:
            status = 0
            xn += 1
            count[xn] = 0
            silence[xn] = 0
            x1[xn] = 0
            x2[xn] = 0
    el = len(x1[:xn])
    if x1[el - 1] == 0:
        el -= 1
    if x2[el - 1] == 0:
        print('Error: Not find endding point!\n')
        x2[el] = fn
    SF = np.zeros(fn)
    NF = np.ones(fn)
    for i in range(el):
        SF[int(x1[i]):int(x2[i])] = 1
        NF[int(x1[i]):int(x2[i])] = 0
    voiceseg = findSegment(np.where(SF == 1)[0])
    vsl = len(voiceseg.keys())
    return voiceseg, vsl, SF, NF


def findSegment(express):
    """
    分割成語音段
    :param express:
    :return:
    """
    if express[0] == 0:
        voiceIndex = np.where(express)
    else:
        voiceIndex = express
    d_voice = np.where(np.diff(voiceIndex) > 1)[0]
    voiceseg = {}
    if len(d_voice) > 0:
        for i in range(len(d_voice) + 1):
            seg = {}
            if i == 0:
                st = voiceIndex[0]
                en = voiceIndex[d_voice[i]]
            elif i == len(d_voice):
                st = voiceIndex[d_voice[i - 1] + 1]
                en = voiceIndex[-1]
            else:
                st = voiceIndex[d_voice[i - 1] + 1]
                en = voiceIndex[d_voice[i]]
            seg['start'] = st
            seg['end'] = en
            seg['duration'] = en - st + 1
            voiceseg[i] = seg
    return voiceseg


def vad_corr(y, wnd, inc, NIS, th1, th2):
    x = enframe(y, wnd, inc)
    Ru = STAc(x.T)[0]
    Rum = Ru / np.max(Ru)
    thredth = np.max(Rum[:NIS])
    T1 = th1 * thredth
    T2 = th2 * thredth
    voiceseg, vsl, SF, NF = vad_forw(Rum, T1, T2)
    return voiceseg, vsl, SF, NF, Rum

```

```python
def enframe(x, win, inc):
    frame_length = len(win)
    num_frames = (len(x) - frame_length) // inc + 1

    frameout = np.zeros((num_frames, frame_length))
    for i in range(num_frames):
        start = i * inc
        end = start + frame_length
        frameout[i, :] = x[start:end] * win

    return frameout

file_path = '../01.wav'
data, fs = librosa.load(file_path, sr=None)
data /= np.max(data)
IS = 0.25
wlen = 200
inc = 80
N = len(data)
time = [i / fs for i in range(N)]
wnd = np.hamming(wlen)
NIS = int((IS * fs - wlen) // inc + 1)
thr1 = 1.1
thr2 = 1.3
voiceseg, vsl, SF, NF, Rum = vad_corr(data, wnd, inc, NIS, thr1, thr2)
fn = len(SF)
frameTime = FrameTimeC(fn, wlen, inc, fs)
plt.figure(figsize=(12,6))
plt.subplot(2, 1, 1)
plt.plot(time, data)
plt.subplot(2, 1, 2)
plt.plot(frameTime, Rum)

for i in range(vsl):
    plt.subplot(2, 1, 1)
    plt.vlines(x=frameTime[voiceseg[i]['start']], ymin=np.min(data), ymax=np.max(data), colors='b', linestyles='dotted')
    plt.vlines(x=frameTime[voiceseg[i]['end']], ymin=np.min(data), ymax=np.max(data), colors='r', linestyles='solid')
    plt.title("纯语音波形")
    plt.ylabel("幅值")
    plt.legend(['波形', '开始', '结束'])

    plt.subplot(2, 1, 2)
    plt.vlines(x=frameTime[voiceseg[i]['start']], ymin=np.min(Rum), ymax=np.max(Rum), colors='b', linestyles='dotted')
    plt.vlines(x=frameTime[voiceseg[i]['end']], ymin=np.min(Rum), ymax=np.max(Rum), colors='r', linestyles='solid')
    plt.title("短时自相关函数")
    plt.ylabel("幅值")
    plt.xlabel("时间/s")
    plt.legend(['短时自相关', '开始', '结束'])
plt.show()

```
