```python
import numpy as np
import scipy.io.wavfile as wav

def SpectralSub(signal, wlen, inc, NIS, a, b):
    # signal: 输入的含噪语音信号
    # wlen: 窗函数或窗长
    # inc: 帧移
    # NIS: 前导无话段帧数
    # a: 过减因子
    # b: 增益补偿因子

    # 将信号切分成帧
    frames = []
    for i in range(0, len(signal) - wlen, inc):
        frame = signal[i:i + wlen]
        frames.append(frame)

    frames = np.array(frames)

    # 计算每帧的功率谱
    mag_frames = np.abs(np.fft.fft(frames, axis=1))
    power_frames = 1.0 / wlen * np.square(mag_frames)

    # 计算噪声估计
    noise_estimation = np.mean(power_frames[:NIS], axis=0)

    # 谱减法降噪
    power_frames_sub = np.maximum(power_frames - a * noise_estimation, 0)

    # 增益补偿
    power_frames_sub_comp = b * power_frames_sub

    # 合成帧
    mag_frames_sub_comp = np.sqrt(power_frames_sub_comp)
    frames_sub_comp = np.fft.ifft(mag_frames_sub_comp, axis=1).real

    # 重叠相加
    output = np.zeros_like(signal)
    for i in range(len(frames_sub_comp)):
        output[i * inc:i * inc + wlen] += frames_sub_comp[i]

    return output

# 测试代码
if __name__ == "__main__":
    # 读取测试语音
    rate, signal = wav.read('../01.wav')

    # 添加不同信噪比的噪声
#     snr = 10 # 信噪比（可以根据需要更改）
    noise = np.random.normal(0, 400, len(signal))
    noisy_signal = signal + noise

    # 调用SpectralSub进行降噪
    output = SpectralSub(noisy_signal, wlen=200, inc=80, NIS=5, a=10, b=1000)

    # 保存降噪结果
    wav.write('output.wav', rate, output.astype(np.int16))

```

```python
def spectral_subtraction(signal, wlen, inc, NIS, a, b):
    # 假设信号是单通道的
    # 分帧
    frames = librosa.util.frame(signal, frame_length=wlen, hop_length=inc).T
    # 窗函数
    window = scipy.signal.get_window('hamming', wlen)
    # 为了保持能量不变，需要对窗函数进行归一化
    window /= np.sum(window**2)**0.5
    # 计算前导无话段的噪声功率谱
    noise_frames = frames[:NIS]
    noise_power = np.mean(np.abs(np.fft.rfft(noise_frames * window, axis=1))**2, axis=0)
    # 初始化输出信号
    output = np.zeros_like(signal)
    # 处理每一帧
    for i, frame in enumerate(frames):
        # 应用窗函数
        windowed_frame = frame * window
        # FFT
        spectrum = np.fft.rfft(windowed_frame)
        # 谱减法
        clean_spectrum = np.maximum(np.abs(spectrum)**2 - a * noise_power, b * noise_power)
        # 保持相位不变
        spectrum = np.sqrt(clean_spectrum) * np.exp(1j * np.angle(spectrum))
        # IFFT
        clean_frame = np.fft.irfft(spectrum)
        # 重叠相加
        output[i*inc:i*inc+wlen] += clean_frame * window
    return output
import  scipy
# 读取测试语音
rate, signal = wav.read('../01.wav')

# 添加不同信噪比的噪声
#     snr = 10 # 信噪比（可以根据需要更改）
noise = np.random.normal(0, 400, len(signal))
noisy_signal = signal + noise

# 调用SpectralSub进行降噪
output = spectral_subtraction(noisy_signal, wlen=200, inc=80, NIS=5, a=9, b=1)

# 保存降噪结果
wav.write('output.wav', rate, output.astype(np.int16))
```
