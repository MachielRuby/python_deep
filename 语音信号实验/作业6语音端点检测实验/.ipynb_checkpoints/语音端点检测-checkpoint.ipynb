```python
# -*- coding:utf-8
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import librosa
import soundfile
import pyaudio #录制和播放
import wave # 导入wave库，处理音频
import time
import matplotlib.pyplot as plt
from scipy.io import wavfile #读取wav提取数据
from scipy.signal import find_peaks
from pydub import AudioSegment #　音频处理
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用SimHei字体（可以根据需要选择其他字体）
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
def play_audio(rate,file_name): # rate 采用率 file_name保存文件名
    Chunk = 1024 # 读取的音频帧的大小
    Format = pyaudio.paInt16 #　音频格式为16位的整数
    Channels = 2 # 通道为2
    Rate = rate #采用率
    Record_times = 5 # 录制时间
    Nframes = int(Rate / Chunk * Record_times) # 计算出所需采集帧的数量
    File_name = file_name
    
    # 实例化一个对象
    p = pyaudio.PyAudio()
    
    # 使用该对象打开声卡，进入参数配置
    stream = p.open(format = Format,
                    channels = Channels,
                    rate = Rate,
                    input = True,
                    frames_per_buffer = Chunk)
    print("开始录音，整个过程持续五秒钟!!!")
    frames = []
    for i in range(0,Nframes):
        data = stream.read(Chunk)
        frames.append(data)
    print("完成录音!!!")
    
    # 关闭
    stream.stop_stream()
    stream.close()
    p.terminate()
    
    # 存储wav
    wf = wave.open(File_name,'wb') # 写入wav文件
    wf.setnchannels(Channels)# 设置通道数
    wf.setsampwidth(p.get_sample_size(Format)) # 设置样本宽度
    wf.setframerate(Rate)# 设置采样率
    wf.writeframes(b''.join(frames)) # 将列表的数据会合写入文件
    wf.close()
    print(f"语音保存为{File_name}")
    
```

```python
# 计算短时能量
def compute_energy(signal, frame_length, frame_step):
    energy = np.array([
        np.sum(signal[i:i+frame_length]**2)
        for i in range(0, len(signal) - frame_length + 1, frame_step)
    ])
    return energy / np.max(energy)

# 计算短时过零率
def compute_zcr(signal, frame_length, frame_step):
    zcr = np.array([
        np.sum(signal[i:i+frame_length-1] * signal[i+1:i+frame_length] < 0)
        for i in range(0, len(signal) - frame_length + 1, frame_step)
    ])
    return zcr
# 双门限法端点检测
def vad_TwoThr(signal, frame_length, frame_step, noise_frames):
    energy = compute_energy(signal, frame_length, frame_step)
    zcr = compute_zcr(signal, frame_length, frame_step)
    
    # 使用前 noise_frames 帧来估计噪声特性
    noise_energy = np.mean(energy[:noise_frames])
    noise_zcr = np.mean(zcr[:noise_frames])
    
    # 设定能量和过零率的阈值
    energy_threshold = 4 * noise_energy
    zcr_threshold = 2 * noise_zcr
    
    # 检测语音活动
    SF = (energy > energy_threshold).astype(int)
    NF = (zcr > zcr_threshold).astype(int)
    
    # 找到语音段的开始和结束
    energy_peaks, _ = find_peaks(energy, height=energy_threshold)
    zcr_peaks, _ = find_peaks(zcr, height=zcr_threshold)
    
    return energy, zcr, SF, NF, energy_peaks, zcr_peaks

# 加载音频文件
file_path = '实验6.wav'
frame_length = 256
frame_step = 80
noise_frames = 10  # 假设前10帧为噪声

# 读取音频文件
sampling_rate, signal = wavfile.read(file_path)

# 确保信号的格式正确
signal = signal / np.power(2, 15) if signal.dtype == np.int16 else signal

# 计算VAD
energy, zcr, SF, NF, energy_peaks, zcr_peaks = vad_TwoThr(signal, frame_length, frame_step, noise_frames)
```
